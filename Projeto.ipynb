```python
import pandas as pd
import pathlib
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
#Importando base de Dados
caminho_base = pathlib.Path('dataset') #Caminho das bases de dados 
meses = {'jan':1, 'fev':2, 'mar':3, 'abr':4, 'mai':5,'jun':6,'jul':7,'ago':8,'set':9,'out':10,'nov':11,'dez':12}
bases = []

for arquivo in caminho_base.iterdir():
    nome_mes = arquivo.name[:3]
    mes = meses[nome_mes]
    ano = arquivo.name[-8:]
    ano = int(ano.replace('.csv',''))

    df = pd.read_csv(caminho_base / arquivo.name)
    df['ano'] = ano
    df['mes'] = mes
    bases.append(df)

base_airbnb = pd.concat(bases)
display(base_airbnb)
```

- Como temos muitas colunas, o modelo pode ficar muito lento.

- Além disso, após uma analise rápida, podemos ver que várias colunas não são necessárias para o nosso modelo, por isso iremos excluir algumas colunas de nossa base de dados.

- Tipos de Colunas que iremos excluir:
    1. IDs, Links e informações não relevantes para o modelo
    2. Colunas repetidas ou extremamente parecidas com outra
    3. Colunas preenchidas com texto livre (Não rodaremos nenhuma ánalise de palavras ou algo do tipo)
    4. Colunas em que todos ou quase todos os valores são iguais

- Depois da ánalise qualitativa das colunas, levando em conta os critérios explicados acima, ficamos com as seguintes colunas:

```python
colunas = ['host_response_time', 'host_response_rate', 'host_is_superhost', 'host_listings_count', 'latitude', 'longitude', 'property_type', 'room_type', 'accommodates', 'bathrooms', 'bedrooms', 'beds', 'bed_type', 'amenities', 'price', 'security_deposit', 'cleaning_fee', 'guests_included', 'extra_people', 'minimum_nights', 'maximum_nights', 'number_of_reviews', 'review_scores_rating', 'review_scores_accuracy', 'review_scores_cleanliness', 'review_scores_checkin', 'review_scores_communication', 'review_scores_location', 'review_scores_value', 'instant_bookable', 'is_business_travel_ready', 'cancellation_policy', 'ano', 'mes']

base_airbnb = base_airbnb.loc[:,colunas]
display(base_airbnb)

```

- Tratamento de dados vazios
- Percebendo a disparidade de dados faltantes, as colunas com mais de 300.000 valores NaN foram excluídas da ánalise

```python
for coluna in base_airbnb:
    if base_airbnb[coluna].isnull().sum()>300000:
        base_airbnb = base_airbnb.drop(coluna,axis=1)

print(base_airbnb.isnull().sum())
```

```python
base_airbnb = base_airbnb.dropna()

print(base_airbnb.shape)
print(base_airbnb.isnull().sum())
```

- Como price e extra_people estão sendo reconhecidos como object ao invés de ser um float, teremos que mudar o tipo da variável.

```python
plt.figure(figsize=(15,10))
sns.heatmap(base_airbnb.corr(numeric_only=True),annot= True, cmap= 'Greens')
```

- Definição de funções para análise de outliers das colunas

```python
def limites(coluna):
    q1 = coluna.quantile(0.25)
    q3 = coluna.quantile(0.75)
    amplitude = q3 - q1
    return (q1 - 1.5 * amplitude, q3+ 1.5 * amplitude)

def diagramaCaixa(coluna):
    fig, (ax1,ax2) = plt.subplots(1,2)
    fig.set_size_inches(15,5)
    sns.boxplot(x = coluna, ax = ax1)
    ax2.set_xlim(limites(coluna))
    sns.boxenplot(x = coluna, ax = ax2)


def histograma(coluna):
    plt.figure(figsize= (15,5))
    sns.distplot(coluna, hist= True)
```

### Price

```python
diagramaCaixa(base_airbnb['price'])
histograma(base_airbnb['price'])
```
